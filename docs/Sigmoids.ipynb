```python
from IPython.html.widgets import interact, interactive, fixed
from IPython.html import widgets
from IPython.display import clear_output, display, HTML

from matplotlib import pyplot as plt

def gauss(x,mu=0,sig=1):
    return np.exp(-(x-mu)**2/(2*sig**2))/(np.sqrt(2*np.pi)*sig)

def sig(p=0.1,W=3):
    figsize(8,5)
    t = np.linspace(-10,10,100)
    Pyx = p * gauss(t,W)
    Px  = p * gauss(t,W) + (1-p) * gauss(t,0)
    PyGx= Pyx/Px
    plt.plot(t,PyGx, lw=0.5, color='blue')
    plt.plot(t,Pyx, lw=2, color='green')
    plt.plot(t,Px, lw=1, color='red')
    plt.axis([-10, 10, 0, 1])    

Res = 200
def sig2D(p=0.1,W1=2,W2=2,s1=1,s2=1.5):    
    x1=np.linspace(-10,10,Res)
    x2=np.linspace(-10,10,Res)
    X1,X2=np.meshgrid(x1,x2)
    
    Px = p * gauss(X1,W1,s1) * gauss(X2,W2,s2) + (1-p) * gauss(X1,0,s1) * gauss(X2,0,s2)
    Pyx = p * gauss(X1,W1,s1) * gauss(X2,W2,s2)
    PyGx = Pyx/Px
    
    plt.figure(figsize=(6,6))
    plt.axis('off')
    plt.pcolormesh(PyGx,cmap="BrBG",alpha=1.0)
    plt.pcolormesh(Pyx,cmap="jet",alpha=0.1)
    plt.pcolormesh(Px,cmap="jet",alpha=0.1)    
    
def nonsig(p1=0.1,p2=0.1,W1=3.3,W2=2):
    figsize(8,5)
    t = np.linspace(-10,10,100)
    P12x = p1 * p2 * gauss(t,W1+W2)
    P10x = p1 * (1-p2) * gauss(t,W1) 
    P20x = p2 * (1-p1) * gauss(t,W2)
    P00x = (1 - p1) * (1 - p2) * gauss(t,0)

    P1x = P12x + P10x
    P2x = P12x + P20x
    Px  = P10x + P20x + P00x + P12x

    P1Gx = P1x/Px
    P2Gx = P2x/Px
    
    plt.plot(t,P1Gx, '-',lw=0.5, color='blue')
    plt.plot(t,P2Gx, '--',lw=0.5, color='blue')
    plt.plot(t,P1x, '-',lw=2, color='green')
    plt.plot(t,P2x, '--',lw=2, color='green')
    plt.plot(t,Px, lw=1, color='red')
    
def nonsig2D(W11=2,W12=2,W21=-2,W22=2):
    p1 = 0.4
    p2 = 0.4    
    x1=np.linspace(-10,10,Res)
    x2=np.linspace(-10,10,Res)
    X1,X2=np.meshgrid(x1,x2)
    
    P12x = p1 * p2 * gauss(X1,W11+W12) * gauss(X2,W21+W22)
    P10x = p1 * (1-p2) * gauss(X1,W11) * gauss(X2,W21) 
    P20x = p2 * (1-p1) * gauss(X1,W12) * gauss(X2,W22)
    P00x = (1 - p1) * (1 - p2) * gauss(X1,0) * gauss(X2,0)
    
    P1x = P12x + P10x
    P2x = P12x + P20x
    Px  = P10x + P20x + P00x + P12x
    
    P1Gx = P1x/Px
    P2Gx = P2x/Px
    
    plt.figure(figsize=(6,6))
    plt.axis('off')
    plt.pcolormesh(P1x,cmap="jet",alpha=0.8)
    plt.pcolormesh(P2x,cmap="jet",alpha=0.8)
    plt.pcolormesh(P1Gx,cmap="BrBG",alpha=0.5)
    plt.pcolormesh(P2Gx,cmap="BrBG",alpha=0.5)    
    plt.pcolormesh(Px,cmap="jet",alpha=0.1)    
```

# **The Ubiquity of Sigmoids**

## Capabilities and Limitations in Neural Networks

(and some mention of the missing data problem)

# Introduction

### Hypothesis Testing and Sigmoids

Suppose we have a binary random variable

$$
\mathbf{Y} = 
\begin{cases}
   1 & \text{with prob. } p \\
   0 & \text{otherwise }
\end{cases}
$$

We cannot observe $\mathbf{Y}$ directly. However, we *can* observe a proxy variable

$$
\mathbf{X} = W\mathbf{Y} + \epsilon
$$
where $\epsilon \sim \mathcal{N}(0,1)$.

Given the observation $\mathbf{X} = \mathbf{x}$, what can we say about $\mathbf{Y}$?

## Hypothesis Testing

In hypothesis testing, we compute the likelihood ratio

$$
R = \frac{\mathbb{P}( \mathbf{X} = \mathbf{x} | \mathbf{Y} = 1)}{ \mathbb{P}( \mathbf{X} = \mathbf{x}|\mathbf{Y} = 0)}
$$



Bayes's Theorem allows us to compute a different quantity:

$$
\mathbb{P}(\mathbf{Y}=1 | \mathbf{X}=\mathbf{x}) = \frac{\mathbb{P}(\mathbf{Y}=1 , \mathbf{X}=\mathbf{x})}{ \mathbb{P}(\mathbf{X}=\mathbf{x})} 
$$

Since 

$$\mathbb{P}(\mathbf{X}=\mathbf{x}) = \mathbb{P}(\mathbf{Y}=1 , \mathbf{X}=\mathbf{x}) + \mathbb{P}(\mathbf{Y}=0 , \mathbf{X}=\mathbf{x})$$

and since
$$
\frac{\mathbb{P}( \mathbf{X} = \mathbf{x}  , \mathbf{Y} = 0)}{\mathbb{P}( \mathbf{X} = \mathbf{x}  , \mathbf{Y} = 1)}
= \frac{\mathbb{P}( \mathbf{X} = \mathbf{x}  | \mathbf{Y} = 0)}
{ \mathbb{P}( \mathbf{X} = \mathbf{x}  | \mathbf{Y} = 1) }
\times 
\frac{\mathbb{P}(\mathbf{Y} = 0)}{\mathbb{P}(\mathbf{Y} = 1)}
$$

we get 

$$
\begin{align}
\mathbb{P}(\mathbf{Y}=1|\mathbf{X}=\mathbf{x}) &= \frac{\mathbb{P}(1,\mathbf{x})}{\mathbb{P}(1,\mathbf{x})+\mathbb{P}(0,\mathbf{x})}\\\\
&= \left(1 + \frac{\mathbb{P}(0,\mathbf{x})  }{ \mathbb{P}(1,\mathbf{x}) } \right)^{-1} \\\\
&= \left(1+ \frac{\mathbb{P}(0)}{\mathbb{P}(1)} \frac{\mathbb{P}(\mathbf{x}|0)  }{ \mathbb{P}(\mathbf{x}|1) } \right)^{-1} \\\\
&= \left(1+ \frac{\mathbb{P}(0)}{\mathbb{P}(1)} \frac{1}{R} \right)^{-1}
\end{align}
$$

## Here comes the sigmoid

In our model, we have
$$
\begin{align}
\mathbb{P}(\mathbf{x}|0) &= \frac{e^{- \frac{\scriptscriptstyle 1}{\scriptscriptstyle 2} \mathbf{x}^2}}{\sqrt{2\pi}} \\\\
\mathbb{P}(\mathbf{x}|1) &= \frac{e^{- \frac{\scriptscriptstyle 1}{\scriptscriptstyle 2} (\mathbf{x}-W)^2}}{\sqrt{2\pi}} 
\end{align}
$$

So that the ratio becomes
$$
\begin{align}
\frac{\mathbb{P}(0)}{\mathbb{P}(1)} \frac{\mathbb{P}(\mathbf{x}|0)  }{ \mathbb{P}(\mathbf{x}|1) } 
&= \frac{1-p}{p} e^{-(W\mathbf{x} - \frac{W^2}{2})} \\\\
&= e^{-(W\mathbf{x} - \frac{W^2}{2} + \log \frac{p}{1-p})}
\end{align}
$$

This finally gives us

$$
\begin{align}
\mathbb{P}(\mathbf{Y}=1| \mathbf{X}= \mathbf{x}) &= \sigma(W\mathbf{x} - \frac{W^2}{2} + \log \frac{p}{1-p}) \\\\
&= \sigma(W\mathbf{x} - b) 
\end{align}
$$

where $\sigma(x) = \frac{1}{1+e^{-x}}$ is the sigmoid-shaped logistic function.

Note that the prior $p$ affects the bias term $b$ as well!

In short: 

$$
\mathbb{P}(\mathbf{Y}=1 | \mathbf{X} =\mathbf{x} ) = \sigma(W\mathbf{x} - b)
$$

where $b$ is a function of $W$ and $p$.

This widget illustrates the effect of $W$ and $p$ on the bias and slope of the sigmoid:

```python
i = interact(sig,p = (0,1,0.01),W=(-5,5,0.1))
```

# Multiple observations

The same holds true if we have multiple observations $\mathbf{x} = (x_1,x_2,\dots,x_N)$ where

$$
x_i = W_i \mathbf{Y} + \epsilon
$$

We still get

$$
\mathbb{P}(\mathbf{Y}=1 | \mathbf{X} =\mathbf{x} ) = \sigma(W\cdot\mathbf{x} - b)
$$

but now $W\cdot\mathbf{x} = \sum_n W_n x_n$.

```python
i = interact(sig2D,p=(0,1,0.1),W1=(-5,5,0.1),W2=(-5,5,0.1),s1=(0,5,0.1),s2=(0,5,0.1))
```

# Missing data

Recalling the original form of $W\mathbf{x} - b$ tells us what we should do when we have missing data.

Following the same derivations for the single observation case, we get the following expression:

$$
\left(\sum_n W_n x_n \right) - \left(\sum_n \frac{1}{2}W_n^2 \right) + \log\frac{p}{1-p} 
$$

If we have fewer than $N$ observations, we simply sum over the observations we have:

$$
\left(\sum_{n \text{ obsv.}} W_n x_n \right) - \left(\sum_{n \text{ obsv.}} \frac{1}{2}W_n^2 \right) + \log\frac{p}{1-p} 
$$

Let $\mathbf{o}$ be a binary vector where

$$
o_n = 
  \begin{cases}
   1 & \text{if sensor } n \text{ is observed} \\
   0 & \text{otherwise }
  \end{cases}
$$

and let the readings we observe be $\mathbf{x}' = \mathbf{x} \star \mathbf{o}$, where $\star$ denotes the elementwise product.

Then in the presence of missing data, our expression takes the form

$$
\mathbf{P}(\mathbf{Y}=1|\mathbf{x}',\mathbf{o}) = \sigma(W\cdot\mathbf{x}' + B\mathbf{o} + b)
$$

Note that this is necessary even when the bias of the noise for each $x_i$ is 0!

# Multiple hidden phenomena

Things get uglier when there are multiple hidden phenomena. Suppose now that our hidden phenomena is a vector $\mathbf{y} = (y_1,\dots,y_M)$ where

$$
y_m = 
\begin{cases}
   1 & \text{with prob. } p_m \\
   0 & \text{otherwise }
\end{cases}
$$

As before, we have multiple observations

$$
\mathbf{x} = W \mathbf{y} + \mathbf{\epsilon},
$$

where $W$ is now an $N \times M$ matrix with entries denoted $W^m_n$.

What does $\mathbb{P}(Y_i = 1| \mathbf{x})$ look like?

```python
i = interact(nonsig,p1 = (0,1,0.01),p2 = (0,1,0.01),W1=(-5,5,0.1),W2=(-5,5,0.1))
```

All is not lost: we *do* recover a sigmoid when $W^m = 0$ for some $m$.

*($W^m$ indicates the $m^{th}$ column of the matrix $W$)*

```python
nonsig(0.1,0.1,2,0)
```

# When do we recover sigmoids?

In the multivariate case, the expression for $\mathbf{P}(\mathbf{x},\mathbf{y})$ is

$$
\frac{1}{\sqrt{2 \pi}}
e^{-\frac{1}{2}||\mathbf{x}||^2}
\left[ \prod_m \left(p_m \cdot e^{W^m\cdot\mathbf{x}} \cdot e^{-\frac{1}{2}||W^m||^2} \right)^{y_m} (1-p_m)^{1-y_m} \right]
\left[ \prod_{m,l} e^{-\frac{1}{2} W^m\cdot W^l \, y_m y_l} \right]
$$



When the columns of $W$ are orthogonal,

$$
W^m\cdot W^l = 0,
$$

we get

$$
\frac{1}{\sqrt{2 \pi}}
e^{-\frac{1}{2}||\mathbf{x}||^2}
\left[ \prod_m \left(p_m \cdot e^{W^m \cdot\mathbf{x}} \cdot e^{-\frac{1}{2}||W^m||^2} \right)^{y_m} (1-p_m)^{1-y_m} \right].
$$

This allows the sum

$$
\mathbb{P}(\mathbf{X} = \mathbf{x}) = \sum_{\mathbf{y}} \mathbb{P}(\mathbf{x},\mathbf{y})
$$

to be factorized

$$
\frac{1}{\sqrt{2 \pi}}
e^{-\frac{1}{2}||\mathbf{x}||^2}
\prod_m \left[p_m \cdot e^{W^m\cdot \mathbf{x} -\frac{1}{2} ||W^m||^2} + (1-p_m) \right]
$$

Focusing on $y_1$, we get

$$
\begin{align}
\mathbb{P}(\mathbf{x},y_1 = 0) &=
\frac{1}{\sqrt{2 \pi}}
e^{-\frac{1}{2}||\mathbf{x}||^2}
\prod_{m>1} \left[p_m \cdot e^{W^m \cdot\mathbf{x} -\frac{1}{2} ||W^m||^2} + (1-p_m) \right]
(1-p_1)
\\\\
\mathbb{P}(\mathbf{x},y_1 = 1) &=
\frac{1}{\sqrt{2 \pi}}
e^{-\frac{1}{2}||\mathbf{x}||^2}
\prod_{m>1} \left[p_m \cdot e^{W^m \cdot\mathbf{x} -\frac{1}{2} ||W^m||^2} + (1-p_m) \right]
p_1 \cdot e^{W^1 \cdot\mathbf{x} -\frac{1}{2} ||W^1||^2} 
\end{align}
$$

Taking the ratio of these quantities, we obtain

$$
\mathbb{P}(y_1 = 1|\mathbf{x}) = \sigma\left(W^1\cdot\mathbf{x} - \frac{1}{2}||W^1||^2 + \log\frac{p}{1-p}\right)
$$

so we recover sigmoids *when our features are orthogonal!*

```python
i = interact(nonsig2D,W11=(-5,5,0.1),W12=(-5,5,0.1),W21=(-5,5,0.1),W22=(-5,5,0.1))
```

Strange things happen when features are not orthogonal:

```python
nonsig2D(-4.2,3,-0.7,-3)
```

## Recovering Orthogonality

Although $W$ may not be orthogonal, there exists a change of basis matrix $\Phi$ such that $\Phi W$ is orthogonal. (In fact, there are infinitely many!)

Let $U = \Phi W$, and 

$$
\begin{align}
\mathbf{z} &= \Phi\mathbf{x} \\\\
&= \Phi (W\mathbf{y} + \epsilon) \\\\
&= U \mathbf{y} + \Phi \epsilon \\\\
&= U \mathbf{y} + \epsilon'.
\end{align}
$$


Each entry of $\epsilon'$ is normally distributed, just like $\epsilon$, so...

$$
\begin{align}
\mathbb{P}(y_i=1|\mathbf{z}) &= \sigma(V^i\cdot \mathbf{z} - \mathbf{b}) \\\\
&= \sigma\left(V^i\cdot(\Phi \mathbf{x}) - \mathbf{b}\right) \\\\
&= \sigma\left((V^i\Phi)\cdot \mathbf{x} - \mathbf{b}\right)
\end{align}
$$

for some $V$ and  $\mathbf{b}$!
